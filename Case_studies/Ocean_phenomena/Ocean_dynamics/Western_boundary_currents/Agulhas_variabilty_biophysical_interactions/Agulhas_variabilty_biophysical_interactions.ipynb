Concurrent with the past century's persistent warming of global oceans, marine heatwaves (periods of extreme regional ocean warming) have become more frequent and more extreme (Laufk√∂tter et al., 2020). They occur in many areas around the world, from the Pacific Ocean to the Atlantic Ocean to the Mediterranean Sea and threaten marine biodiversity and its ecosystems (Smale et al., 2019). One particular ecosystem impacted by marine heatwaves, are coral reefs. These heatwaves can, for example, cause coral bleaching, coral disease outbreaks, and/or algae blooms (Roberts et al., 2019). Although coral can survive these so-called bleaching events, they become more stressed, more susceptible to diseases, and, on the long-term, subject to mortality (NOAA, 2021).

In this notebook we will work through an example of how you can access near-real-time data to view current SST in a region that is often affected by marine heatwaves. We will then look at this area in a long term context using a reprocessed time series, to see how the current situation compares to historical marine heat wave episodes.

This Jupyter Notebook builds on the **<a href="https://www.eumetsat.int/Marine-heatwave-intensification-threatens-coral-reef-health" target="_blank">Marine heatwave intensification threatens coral reef health</a>** case study, and will replicate the figures 3, 5 and 6 using Level-2 data from the Copernicus Sentinel-3 SLSTR sensor and Level-4 data from the Level-4 global OSTIA record. Before we start the notebook proper, we will define a quick function that we will use throughout. The function is quite simple and just adds embellishments (gridlines, labels etc) to plots.

````python
def embellish_plot(m):
    """Quick function to embellish SLSTR plots with gridlines and labels

    Args:
        m (axis): the axis to plot into
    Returns:
        None
    """
    # Embellish with gridlines
    g1 = m.gridlines(draw_labels = True, zorder=20, color='0.0', linestyle='--',linewidth=0.5)
    g1.xlocator = mticker.FixedLocator(np.arange(-180, 180, 2))
    g1.ylocator = mticker.FixedLocator(np.arange(-90, 90, 2))
    g1.top_labels = False
    g1.right_labels = False
    g1.xlabel_style = {'color': 'black'}
    g1.ylabel_style = {'color': 'black'}
    m.set(facecolor = "0.75")
    #m.axis('off')
````

````python
def process_SRAL_track(SRAL_var, plot_region, cwin=1, adt=True):
    """Function to extract geostrophic velocity components from SRAL passes
    
    Args:
        SRAL_var (dataset)          : SRAL product as xarray dataset
        track_direction (integer)   : 0/1 (ascending/descending)
        plot_region (list)          : subset region [W,E,S,N]
    Returns:
        sla, lon, lat, u, v (array) : sea level anomaly, lat, lon, u-geostrophic, v-geostrophic 
    """
              
    # subset
    if adt:
        ssh_var = SRAL_var.ssha_01_ku.data + SRAL_var.mean_dyn_topo_01.data
        label = "Absolute dynamic topography"
    else:
        ssh_var = SRAL_var.ssha_01_ku.data
        label = "Sea level anomaly"

    ssh_var[SRAL_var.surf_type_01.data != 0.0 ] = np.nan

    ds = xr.Dataset({'ssha01': (['time'], ssh_var)}, coords={'lon01': (['time'], SRAL_var.lon_01.data), 'lat01': (['time'], SRAL_var.lat_01.data)})
    extr = ds.sel(time=(ds.lat01 > plot_region[2] - pad) & (ds.lat01 < plot_region[3] + pad))
    sla = extr['ssha01']
    lon = extr['lon01']
    lat = extr['lat01']

    #earth radius in meters (roughly, on average : 6371 km)
    earth_radius = 6371000
    # gravitational constant
    g = 9.81
    # Coriolis frequency f = 2 * Earth rotation rate * sin(lat). !!! not at the Equator !!!
    f = 2 * (7.2921*10**-5) * np.sin(np.radians(lat))
    f[f == 0.0] = np.nan
    
    # calculate angular and sla deltas
    dLon = np.radians(lon[cwin:] - lon[:-cwin])
    dLat = np.radians(lat[cwin:] - lat[:-cwin])
    dH = sla[2*cwin:] - sla[:-2*cwin]
    dH = dH.pad(time=(0,cwin))

    # distance in meters along longitude ending at i (NB. dLon in radians) 
    dx_before = earth_radius * dLon * np.cos(np.radians(lon[cwin:]))
    # distance in meters along longitude ending at i+1 (NB. dLon in radians)
    dx_after = earth_radius * dLon * np.cos(np.radians(lat[:-cwin]))
    # distance in meters along latitude, between i+1 and i (NB. dLat in radians)
    dy = earth_radius * dLat
    
    distance = np.sqrt(dx_before ** 2 + dy ** 2) + np.sqrt(dx_after ** 2 + dy ** 2)
    distance[distance == 0.0] = np.nan
    
    if np.array(lat)[-1] > np.array(lat)[0] > 0: # ascending
        direction = 1
        angle = np.arctan2(dy, dx_before)
    else:                   # descending
        direction = -1
        angle = np.arctan2(dy, dx_before) + np.pi

    # to be used only if distance is not zero, and f neither.
    velocity = direction * (-g / f[:-cwin]) * (dH) / distance

    return np.array(lon), np.array(lat), np.array(sla), np.array(velocity*np.cos(angle - np.pi/2)), np.array(velocity*np.sin(angle - np.pi/2)), label
````

````python
OLCI_dict = {}
for downloaded_file, granule in zip(downloaded_files["OLCI"], range(len(downloaded_files["OLCI"]))):
    OLCI_vars = xr.open_mfdataset(glob.glob(os.path.join(downloaded_file, '*.nc')))

    # subset OLCI from half orbit for ease of plotting
    y, x, mask = eumartools.subset_image(np.array(OLCI_vars["longitude"]), np.array(OLCI_vars["latitude"]),
                                    [plot_region[0]-pad, plot_region[1]+pad, plot_region[1]-pad, plot_region[0]+pad],
                                    [plot_region[2]-pad, plot_region[2]-pad, plot_region[3]+pad, plot_region[2]+pad])

    # define mask
    flag_file = glob.glob(os.path.join(downloaded_file, 'wqsf.nc'))[0]
    NN_mask = eumartools.flag_mask(flag_file, 'WQSF', ['LAND', 'CLOUD', 'CLOUD_AMBIGUOUS', 'CLOUD_MARGIN', 
                                                        'INVALID', 'COSMETIC', 'SATURATED', 'SUSPECT',
                                                        'SNOW_ICE', 'OCNN_FAIL'])
    
    NN_mask = NN_mask[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce].astype(float)
    NN_mask[NN_mask != 0.0] = np.nan
    NN_mask[NN_mask == 0.0] = 1.0

    OLCI_dict[granule] = {'lon': np.array(OLCI_vars.longitude[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]),
                          'lat': np.array(OLCI_vars.latitude[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]),
                          'chl': np.array(OLCI_vars.CHL_NN[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]*NN_mask),
                          'tsm': np.array(OLCI_vars.TSM_NN[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]*NN_mask)}
    OLCI_vars.close()
````

Process SLSTR

````python
SLSTR_dict = {}
for downloaded_file, granule in zip(downloaded_files["SLSTR"],range(len(downloaded_files["SLSTR"]))):
    SLSTR_vars = xr.open_mfdataset(glob.glob(os.path.join(downloaded_file, '*.nc')))
    
    # subset SLSTR from full orbit for ease of plotting
    y, x, mask = eumartools.subset_image(np.array(SLSTR_vars["lon"]), np.array(SLSTR_vars["lat"]),
                                    [plot_region[0]-pad, plot_region[1]+pad, plot_region[1]-pad, plot_region[0]+pad],
                                    [plot_region[2]-pad, plot_region[2]-pad, plot_region[3]+pad, plot_region[2]+pad])

    SST = np.array(SLSTR_vars.sea_surface_temperature[0, min(x):max(x):grid_reduce, min(y):max(y):grid_reduce] +
                                  SLSTR_vars.sses_bias[0, min(x):max(x):grid_reduce, min(y):max(y):grid_reduce] - 273.15)
    QC = np.array(SLSTR_vars.quality_level[0, min(x):max(x):grid_reduce, min(y):max(y):grid_reduce])
    SST[QC < 4] = np.nan

    SLSTR_dict[granule] = {'lon': np.array(SLSTR_vars.lon[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]),
                           'lat': np.array(SLSTR_vars.lat[min(x):max(x):grid_reduce, min(y):max(y):grid_reduce]),
                           'sst': SST}
    SLSTR_vars.close()
````

Process SRAL passes for cross track geostrophic velocity

````python
cwin = 1
SRAL_dict = {}
for downloaded_file, track in zip(downloaded_files["SRAL"], range(len(downloaded_files["SRAL"]))):
    SRAL_vars = xr.open_mfdataset(glob.glob(os.path.join(downloaded_file, 'standard_measurement.nc'))) 
    lon, lat, sla, u, v, label = process_SRAL_track(SRAL_vars, plot_region, cwin=cwin)
    spd = (u**2 + v**2)**0.5
    spd[spd>3]=np.nan
    SRAL_dict[track] = {'lon':lon, 'lat':lat, 'sla':sla, 'u':u, 'v':v, 'spd':spd, 'label':label}
````

````python
fig, m = plt.subplots(1, 1, figsize=(20, 15), dpi=150, subplot_kw={"projection": cartopy.crs.PlateCarree()})

vmin = 12
vmax = 27

# Plot bathymetry
if plot_bathymetry:
    m.contour(blon, blat, bathy, bcontours, linewidths=1.0, linestyles='dashed',
              colors='k', zorder=2)

# Plot SLSTR
for granule in SLSTR_dict:
    c1 = m.contourf(SLSTR_dict[granule]['lon'], SLSTR_dict[granule]['lat'], 
                    SLSTR_dict[granule]['sst'], np.arange(vmin, vmax, 0.5), vmin=vmin, vmax=vmax,
                    cmap=plt.cm.RdYlBu_r, zorder=1)

# Plot altimetry tracks and velocity vectors
for track in SRAL_dict:

    m.scatter(SRAL_dict[track]["lon"], SRAL_dict[track]["lat"], c="0.25", alpha=1, s=200, marker='o', linewidth=0.01, zorder=3)
    
    c2 = m.scatter(SRAL_dict[track]["lon"], SRAL_dict[track]["lat"], c=SRAL_dict[track]["sla"],
                   vmin=np.nanmax(np.abs(SRAL_dict[track]["sla"]))*-1, vmax=np.nanmax(np.abs(SRAL_dict[track]["sla"])),
                   alpha=1, cmap=plt.cm.RdBu_r, s=100, marker='o', linewidth=0.01, zorder=4)

    c3 = m.quiver(SRAL_dict[track]["lon"][:-cwin], SRAL_dict[track]["lat"][:-cwin],
                  SRAL_dict[track]["u"], SRAL_dict[track]["v"], SRAL_dict[track]["spd"],
                  cmap=plt.cm.Purples, headwidth=0, headlength=0, clim=(0, 2),
                  headaxislength=0, scale=20, width=0.002, zorder=5, alpha=0.75, transform=cartopy.crs.PlateCarree())

# Embellish
m.set_extent(plot_region, crs=cartopy.crs.PlateCarree())
m.add_feature(cartopy.feature.LAND, zorder=10, edgecolor='k', facecolor='#5D6D7E', linewidth=0.5)
embellish_plot(m)

# Add colour bars
cb_ax = fig.add_axes([m.get_position().x0, 0.15, m.get_position().width*0.475, 0.01])
cbar = fig.colorbar(c1, cax=cb_ax, orientation="horizontal", label="Sea surface temperature [$^{o}C$]")
cbar.set_ticks(np.arange(vmin, vmax))

cb_ax = fig.add_axes([m.get_position().x0 + m.get_position().width*0.525, 0.15, m.get_position().width*0.475, 0.01])
cbar = fig.colorbar(c2, cax=cb_ax, orientation="horizontal", label=f"{SRAL_dict[track]['label']} [$m$]")

cb_ax = fig.add_axes([m.get_position().x0 + m.get_position().width*0.525, 0.075, m.get_position().width*0.475, 0.01])
cbar = fig.colorbar(c3, cax=cb_ax, orientation="horizontal", label="Cross track geostrophic speed [$ms^{-1}$]")

plt.savefig('SST_SLA_Geostrophic_velocity.png', bbox_inches="tight")
plt.show()
````
