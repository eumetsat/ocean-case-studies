### Outline

El Niño-Southern Oscillation (ENSO) is a natural coupled ocean-atmospheric pattern occurring in the tropical Pacific, but with global impacts. Its cycle ranges between 2 to 7 years causing a variation in temperature of the upper ocean. While El Niño describes the unusual warming of surface waters, La Niña represents the opposite effect and it is often reported as the “cool phase”. The phenomenon was first recorded through observations made by Peruvian fisherman in December during the 18th century. Because of its correspondence with the Christian holy day Christmas, the event was named after the Christ child, literally “the little boy” or “El Niño” in Spanish. Early in the 19th century, the ocean signal was found to correspond with atmospheric events on the other side of the tropical Pacific Ocean - know as the “Southern Oscillation”. ENSO has an extensive influence on regional temperature, precipitation and weather patterns, and effects ocean productivity and fisheries by changing the dynamics of upwelling systems. Consequently, ENSO has attracted significant interest from the scientific community and government agencies, who seek to understand both the drivers and ramifications of its more extreme phases. Strong El Niño events, which are associated with vast amount of heat rising from the Pacific waters and often causing the hottest years in records, are increasing in frequency due to climate change [1], exacerbating our need to monitor and respond to them.

With the advent of satellites technologies, the scientific knowledge regarding the wide-range effects of ENSO on the planet has improved significantly. Earth observation gives us the tools to track the development and effects of the phases of ENSO globally, giving us an up-to-date view across the ocean, land and in the atmosphere. Ocean-observing satellites such as Sentinel-3 and Sentinel-6 give us detailed insights into changes in the sea-surface height (via the SRAL and Poseidon-4 altimeters), sea surface temperature (SLSTR) and ocean colour (OLCI). These observations are essential inputs to downstream products, which integrate information from multiple sensors, and give us the best view of the evolving ENSO situation. These downstream products include those offered by the Copernicus Marine Service, which we use here.

In this Jupyter Notebook we will work through an example of how you can access Level-4 sea level anomaly data from the Copernicus Marine Service. We will show how to manipulate the data to give us weekly spatial anomalies over our area of interest. Finally, we will show how to make and interpret a "Hovmoller" plot of time vs longitude to track these sea level anomalies as the move across the tropical Pacific Ocean. This Jupyter Notebook builds on the <a href="ADD LINK!" target="_blank">Observing marine ENSO progression throughout 2023</a> case study, and will replicate figure 1 from it.

As part of the <a href="https://www.oceandecade.org/" target="_blank"> United Nations Ocean Decade</a>, ten specific <a href="https://oceandecade.org/challenges/
" target="_blank">challenges</a> are being addressed. This work, and the data underlying it, support Challenge 5 - Unlock ocean-based solutions to climate change. Long-term satellite records, such as those from the altimetry missions, allows us to monitoring the evolution of interannual to decadal climate modes such as ENSO and provide us with essential information on the how these patterns are evolving under climate change and insight into the global consequences of the extremes that they bring.

## <a id='section2'></a>2. Downloading level-4 altimetry data
[Back to top](#TOCTOP)

We are going to work with the level-4 sea level anomaly altimetry product from the Copernicus Marine Service Data Store. Level-4 altimetry products contain data from multiple altimeters, including Sentinel-3A and 3B SRAL and Sentinel-6. However, while level-2 products contain along-track data from individual instruments, and level-3 products contain along-track data from multiple instruments, the level-4 merges the level-3 product using optimally interpolation. This results in a fully gridded field. You can find more informaion on this process via the links under the Data Used section, above.

Now that we have anuthenticated the API, we can retrieve our data. The cell below will do that, but note that we had to define some key pieces of information first;
* we need to specify the limits of our region interest (minimum_logitude etc.)
* we need to specify our time bounds in the correct format (start_datetime / end_datetime)
* and we need to specify the dataset_id and variables we want.

The **dataset_id** is a little complex to find as it is not the same as the product ID we reference above. Many products contain multiple datasets, and we need to reference the correct one. You can find the datasets under each product by using the <a href="https://data.marine.copernicus.eu/product/SEALEVEL_GLO_PHY_L4_NRT_008_046/download" target="_blank">"download" service</a> link. This link will also tell you which **variables** are available, and what they are called. The variables should be stored as strings in a comma separated list, e.g. ["var1", "var2" "var3"], as below.

Once we have populate all this information, we can run the cell below to prepare and retrieve our data.

```python
ds = copernicusmarine.open_dataset(
   dataset_id = "cmems_obs-sl_glo_phy-ssh_nrt_allsat-l4-duacs-0.25deg_P1D",
   minimum_longitude = 120,
   maximum_longitude = 280,
   minimum_latitude = -2,
   maximum_latitude = 2,
   start_datetime = "2022-01-01 00:00:00",    
   end_datetime = "2024-02-29 23:59:59", 
   variables = ['sla'], 
)
```

Now we should have all of our data ready to use and stored in the "ds" `xarray.Dataset` object. Lets see what we have.....

```python
ds
```

You should see that you have an `xarray.Dataset` with `Dimensions` of **time**, **latitude** and **longitude** also containing the **sla** `Data variable`. ## <a id='section3'></a>3. Preparing the data
[Back to top](#TOCTOP)

</div>

Now we have our data, our preparation steps are very simple (thanks to the magic of xarray!). First, we are going to average our dataset average all latitudes. This will give us the two dimensional record of longitude vs time that we need for our Hovmoller plot.

```python
# average over the latitudes without NaNs
nclatmean = ds.mean(dim='latitude', skipna=True)
```

Now we are going to resample the time dimension so that we have weekly mean values. We do this to smooth our data a little as are not interested in the changes in our signal on very short timescales.

```python
# resample weekly
nclatmean_weekly = nclatmean.resample(time='W', skipna=True).mean()
```

<div class="alert alert-info" role="alert">

## <a id='section4'></a>4. Making a Hovmoller plot
[Back to top](#TOCTOP)

</div>

Now we are ready too plot our data. First, to set up our 2-dimensional axes, we will create the time2 and lon2 from our 1-dimensional longitude and time coordinates.

```python
# create a longitude-time grid to plot.
time2, lon2 = np.meshgrid(nclatmean_weekly.longitude, nclatmean_weekly.time)
```

Next, we will define the limits our of colour bar, as below.

```python
# Here, we define min / max of the color bar. We shift the colorbar middle 
# in order to balance colors in the plot since sla are referenced to a 20-year
# average, not the current 30+ years available
colormax = np.round(np.nanmax(nclatmean_weekly.sla), 2)
colormin = np.round(np.nanmin(nclatmean_weekly.sla), 2)
```

Now lets set up our figure and make our plot. We will also save it the figure.

```python
# set up our figure
fig = plt.figure(figsize=(35,25))

# set up our axes
ax = fig.add_subplot()
ax.invert_yaxis()

# plot out main panel of data; the Hovmoller plot
mm = ax.pcolormesh(time2, lon2, nclatmean_weekly.sla, vmin=colormin, vmax=colormax, cmap='RdYlBu_r')

# label and tidy up our plot
ax.grid(color='gray', linestyle='--', linewidth=2)
ax.set_xlabel("Longitude")

# plot our colorbar
cbar = fig.colorbar(mm, orientation='horizontal', pad=0.08, shrink=0.66, aspect=36)
cbar.set_label('Sea Level Anomaly [m]', fontsize=36)

# save our plot
plt.savefig('hovmoller_ENSO_2022-2024.png', dpi=150, bbox_inches='tight', pad_inches=0.01)
```

We have now created our Hovmoller plot! Hovmoller plots are a common way to visualise oceanographic data and effectively shows the movement of a signal, such as a wave, on a single spatial axis (in our case longitude) over time. 

During El Niño events, a weakening of the prevailing east-west trade winds gives rise to a characteristic increase in sea level (in red around January 2023 on the left hand side of our plot, around 120°E), which travels across the tropical Pacific basin west to east from Indonesian waters to the coast of Central and South America by around July 2023. The impact of the La Niña from the previous year (2022), can also be seen in the lower sea level on the eastern coasts of the basin (the dark blue are in January 2022 around 240°E-280°E).
